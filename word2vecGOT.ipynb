```python
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
story = []
data_dir = '/content/got'

for filename in os.listdir(data_dir):
    if filename.endswith('.txt'):
        with open(os.path.join(data_dir, filename), 'r', encoding='utf-8', errors='ignore') as f:
            corpus_text = f.read()
            raw_sent = sent_tokenize(corpus_text)
            for sent in raw_sent:
                story.append(simple_preprocess(sent))

print(f'Total sentences in corpus: {len(story)}')
```

```python
print('Sample from corpus:')
display(story[:2])
#In Python, [:2] is a slicing operator.

# When applied to a list like story[:2], it means: "give me all elements from the start of the list up to (but not including) index 2."
#So the 1st two preprocessed sentences will be displayed .
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    vector_size=100,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('stark')
```

```python
model.wv.most_similar('tyrion')
```
